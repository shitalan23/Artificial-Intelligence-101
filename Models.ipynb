```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Parameters for the cubic model
a = -0.5  # Cubic coefficient
b = -1.5  # Quadratic coefficient
c = 2    # Linear coefficient
d = 3    # Intercept
sigma = 10   # Standard deviation of the error term

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, 100).reshape(-1, 1)
error = np.random.normal(0, sigma, size=len(x))
y = a * x.ravel()*3 + b * x.ravel()*2 + c * x.ravel() + d + error
```

```python



import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(-10, 10, 100)
y = 0.5 * x**3 - 2 * x**2 + 3 * x + np.random.normal(0, 10, size=x.shape)

underfit_degree = 1
properfit_degree = 3
overfit_degree = 29

underfit_coeffs = np.polyfit(x, y, underfit_degree)
properfit_coeffs = np.polyfit(x, y, properfit_degree)
overfit_coeffs = np.polyfit(x, y, overfit_degree)

y_underfit = np.polyval(underfit_coeffs, x)
y_properfit = np.polyval(properfit_coeffs, x)
y_overfit = np.polyval(overfit_coeffs, x)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].scatter(x, y, color='black', label='Data', alpha=0.6)
axes[0].plot(x, y_underfit, label=f'Underfit (Degree {underfit_degree})', color='red', linestyle='dashed')
axes[0].set_title(f'Underfit (Degree {underfit_degree})')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[0].grid(True)

axes[1].scatter(x, y, color='black', label='Data', alpha=0.6)
axes[1].plot(x, y_properfit, label=f'Proper Fit (Degree {properfit_degree})', color='green', linestyle='solid')
axes[1].set_title(f'Proper Fit (Degree {properfit_degree})')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].legend()
axes[1].grid(True)

axes[2].scatter(x, y, color='black', label='Data', alpha=0.6)
axes[2].plot(x, y_overfit, label=f'Overfit (Degree {overfit_degree})', color='blue', linestyle='dotted')
axes[2].set_title(f'Overfit (Degree {overfit_degree})')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(scale=0.1, size=x.shape)  # Sinusoidal data with noise

# Fit parameters
degrees = [1, 3, 10]  # Underfit, proper fit, overfit

# Create a figure for plotting
plt.figure(figsize=(15, 5))

# Loop through the degrees and plot the fits
for i, degree in enumerate(degrees):
    # Fit a polynomial of the specified degree
    coeffs = np.polyfit(x, y, degree)
    poly = np.poly1d(coeffs)

    # Generate fitted values
    y_fit = poly(x)

    # Plot the original data and the fitted curve
    plt.subplot(1, 3, i + 1)
    plt.scatter(x, y, color='blue', label='Data', s=10)
    plt.plot(x, y_fit, color='red', label=f'Polynomial Degree {degree}')
    plt.title(f'Degree {degree} Fit')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim(-2, 2)  # Set y-limits for better visualization
    plt.legend()
    plt.grid()

# Show the plots
plt.tight_layout()
plt.show()
```
